```python
import numpy as np
import matplotlib.pyplot as plt

from torchvision import datasets, transforms
```

```python
# Define Transformer to turn images to tensors
transform = transforms.ToTensor()
```

```python
# Download and set the dataset
train_dataset = datasets.MNIST(
    root='./data', # Create 'data' File
    train=True, # All examples to training
    download=True, # Download from internet
    transform=transform # Transform to tensors
)
```

```python
test_dataset = datasets.MNIST(
    root='./data',
    train=False,
    download=True,
    transform=transform
)
```

```python
print(f"Training images: {len(train_dataset)}")
print(f"Testing images: {len(test_dataset)}")
```

```python
from torchvision.transforms import ToPILImage
transform = ToPILImage()
```

```python
image = transform(train_dataset[0][0])
image.show()
```

```python
fig, axes = plt.subplots(2,5,figsize=(10,5))

for i, ax in enumerate(axes.flat):
    image, label = train_dataset[i]

    image_np = image.squeeze().numpy()

    ax.imshow(image_np, cmap='gray')
    ax.set_title(label)
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
sevens = (train_dataset.targets == 7).nonzero(as_tuple=True)[0]

fig, axes = plt.subplots(2,6,figsize=(15,6))

for i, ax in enumerate(axes.flat):

    idx = sevens[i].item()
    image, label = train_dataset[idx]
    image_np = image.squeeze().numpy()

    ax.imshow(image_np, cmap='gray')
    ax.axis('off')

plt.tight_layout()
plt.show()
```
