```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

from IPython import display
```

## Create data

```python
N = 30
x = torch.randn(N, 1)
y = x + torch.randn(N,1)

plt.plot(x, y, 's')
```

## Build Model

```python
ANNreg = nn.Sequential(
    nn.Linear(1,1),  # input layer
    nn.ReLU(),       # activation function
    nn.Linear(1,1)   # output layer
)

ANNreg
```

## Parameters

```python
learningRate = 0.05

lossFun = nn.MSELoss()

optimizer = torch.optim.SGD(ANNreg.parameters(), lr=learningRate)
```

## Train the model

```python
numEpochs = 100
losses = torch.zeros(numEpochs)
count_losses = 0
for epochi in range(numEpochs):

    yPred = ANNreg(x)

    loss = lossFun(yPred, y)
    losses[epochi] = loss

    if loss.item() < 0.25:
        count_losses += 1

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
count_losses
```

## Test the model

```python
predictions = ANNreg(x)

test_loss = (predictions-y).pow(2).mean()

plt.plot(losses.detach(), 'o', markerfacecolor='w', linewidth=.1)
plt.plot(numEpochs, test_loss.detach(), 'ro')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Final loss = %g' %test_loss.item())
plt.show()
```
