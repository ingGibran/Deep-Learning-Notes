```python
import numpy as np
import torch
import torch.nn as nn
```

```python
widenet = nn.Sequential(
    nn.Linear(2,4),
    nn.Linear(4,3)
)

deepnet = nn.Sequential(
    nn.Linear(2,2),
    nn.Linear(2,2),
    nn.Linear(2,3)
)

print(widenet)
print(' ')
print(deepnet)
```

```python
for p in deepnet.named_parameters():
    print(p)
    print(' ')
```

```python
for p in widenet.parameters():
    print(p)
    print(' ')
```

```python
numNodesInWide = 0
for p in widenet.named_parameters():
    if 'bias' in p[0]:
        numNodesInWide += len(p[1])

numNodesInDeep = 0
for paramName, paramVect in deepnet.named_parameters():
    if 'bias' in paramName:
        numNodesInDeep += len(paramVect)

print("Wide:", numNodesInWide)
print("Deep:", numNodesInDeep)
```

```python
nparams = 0
for p in widenet.parameters():
    if p.requires_grad:
        print('Parameters: ', p.numel())
        nparams += p.numel()

print("Total: ", nparams)
```

```python
nparams = np.sum([p.numel() for p in widenet.parameters() if p.requires_grad ])
print("widenet: ", nparams)

nparams = np.sum([p.numel() for p in deepnet.parameters() if p.requires_grad ])
print("Deepnet: ", nparams)
```

```python
from torchsummary import summary

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
widenet.to(device)

summary(widenet, (1,2))
```
