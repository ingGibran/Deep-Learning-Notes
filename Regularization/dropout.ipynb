```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
prob = .25

dropout = nn.Dropout(p=prob)

x = torch.ones(10)

y = dropout(x)
print(x)
print(y)
print(torch.mean(y))

nodes = []
for i in range(10):
    if y[i] == 0:
        nodes.append('0')
    else:
        nodes.append('Node')
print(nodes)
```

```python
dropout.eval()

y = dropout(x)

print(y)
print(torch.mean(y))
```

```python
y = F.dropout(x, training=False) # bool -> self.istraining

print(y)
print(torch.mean(y))
```

```python
dropout.train()
y = dropout(x)
print(y)

dropout.eval()
y = dropout(x)
print(y)

#dropout.train()
y = dropout(x)
print(y)
```
