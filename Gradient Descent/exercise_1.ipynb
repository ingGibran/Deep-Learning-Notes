## Gradient Descent

```python
import numpy as np
```

```python
localmin = np.array([0])

learning_rate = 0.01
training_epochs = 100

for i in range(training_epochs):
    grad = diff(localmin, x)
    localmin = localmin - learning_rate*grad

localmin
```
