## Tabla for Scikit-Learn

```python
table = df.drop(columns="loan_id")
```

```python
table["education"] = table["education"].replace({" Graduate": 1, " Not Graduate": 0})
table["self_employed"] = table["self_employed"].replace({" Yes": 1, " No": 0})
table["loan_status"] = table["loan_status"].replace({" Approved": 1, " Rejected": 0})
```

```python
df["loan_status"] = df["loan_status"].astype(int)
```

```python
print(table["education"].value_counts())
print(table["self_employed"].value_counts())
print(table["loan_status"].value_counts())
```

## Divide table into features and label

```python
X = table.drop(columns="loan_status")
y = table["loan_status"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=77)
```

## Normalize Features

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

vars = [
    "no_of_dependents",
    "income_annum", 
    "loan_amount", 
    "loan_term", 
    "cibil_score", 
    "residential_assets_value", 
    "commercial_assets_value", 
    "luxury_assets_value", 
    "bank_asset_value"
]

X_train[vars] = scaler.fit_transform(X_train[vars])
X_test[vars] = scaler.fit_transform(X_test[vars])
X_train
```

```python
table["loan_status"].unique()
```

## Implement Scikit-Learn Model

### Linear Regression

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
from sklearn.metrics import classification_report

classification_report(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```
