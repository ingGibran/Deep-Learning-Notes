# Mean & Variance

## numpy

```python
import numpy as np
```

```python
x = [1, 2, 4, 6, 5, 4, 0]
n = len(x)

mean1 = np.mean(x)
mean2 = np.sum(x) / n

print(mean1)
print(mean2)
```

```python
var1 = np.var(x)
var2 = np.var(x, ddof=1)
var3 = (1/(n-1)) * np.sum((x-mean1)**2)

print(var1)
print(var2)
print(var3)
```
