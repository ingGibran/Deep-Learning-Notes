# Derivatives

```python
import numpy as np
import sympy as sym

from IPython.display import display
```

```python
x = sym.symbols('x')

fx = 2*x**2
gx = 4*x**3 - 3*x**4

df = sym.diff(fx)
dg = sym.diff(gx)

display(df)
display(dg)
print()

# WRONG
display(df*dg)

# RIGHT
mult = sym.diff(fx*gx)
display(mult)

```

```python
fx = (x**2 + 4*x**3)**5

print('Function: ')
display(fx)
print()

print('Derivate: ')
display(sym.diff(fx))
    
```
