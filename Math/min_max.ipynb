# Min/Max & Argmin/Argmax

```python
import numpy as np
import torch
import torch.nn as nn
```

## numpy

```python
v = np.array([1, 40, 2, -3])

minval = np.min(v)
maxval = np.max(v)

print(f"Min: {minval}, Max: {maxval}")

minidx = np.argmin(v)
maxidx = np.argmax(v)

print(f"Argmin: {minidx}, Argmax: {maxidx}")
```

```python
M = np.array([
    [0, 1, 10],
    [20, 8, 5]
])

print(M), print()

minvals1 = np.min(M)
minvals2 = np.min(M, axis=0)
minvals3 = np.min(M, axis=1)

print(minvals1)
print(minvals2)
print(minvals3)
```

```python
minidx1 = np.argmin(M)
minidx2 = np.argmin(M, axis=0)
minidx3 = np.argmin(M, axis=1)

print(M), print()
print(minidx1)
print(minidx2)
print(minidx3)
```

## pytorch

```python
v = torch.tensor([1, 40, 2, -3])

minval = torch.min(v)
maxval = torch.max(v)

print(f"Min: {minval}, Max: {maxval}")

minidx = torch.argmin(v)
maxidx = torch.argmax(v)

print(f"Argmin: {minidx}, Argmax: {maxidx}")
```

```python
M = torch.tensor([
    [0, 1, 10],
    [20, 8, 5]
])

print(M), print()

min1 = torch.min(M)
min2 = torch.min(M, axis=0)
min3 = torch.min(M, axis=1)

print(min1), print()
print(min2), print()
print(min2.values), print()
print(min2.indices), print()
print(min3)
```
