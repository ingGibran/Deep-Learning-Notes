# Transpose Vector and Matrix

## Imports

```python
import numpy as np
import torch
```

## Using Numpy


```python
# create a vector
nv = np.array([ [1, 2, 3, 4] ])
print(nv), print(' ')

# transpose it
print(nv.T), print(' ')

# transpose the transpose
nvT = nv.T
print(nvT.T)
```

```python
# repeat for a matrix
nM = np.array([
    [1, 2, 3, 4],
    [5, 6, 7, 8]
])
print(nM), print(' ')

# transpose it
print(nM.T), print(' ')

# transpose the transpose
nMT = nM.T
print(nMT.T)
```

## Using PyTorch

```python
# create a vector
tv = torch.tensor([ [1, 2, 3, 4] ])
print(tv), print(' ')

# transpose it
print(tv.T), print(' ')

# transpose the transpose
tvT = tv.T
print(tvT.T)
```
