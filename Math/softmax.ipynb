# Softmax

```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

## numpy

```python
z = np.random.randint(-5, high=15, size=100)

num = np.exp(z)
den = np.sum(np.exp(z))
sigma = num / den

print(sigma)
print(np.sum(sigma))

plt.plot(z,sigma,'ko')
plt.xlabel('Original number (z)')
plt.ylabel('Softmaxified $\sigma$')
plt.show()
```
