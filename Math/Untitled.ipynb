# Random

## numpy

```python
import numpy as np
```

```python
np.random.randint(50)
```

```python
np.random.seed(17)
np.random.randint(50)
```

```python
randseed1 = np.random.RandomState(17)
randseed2 = np.random.RandomState(200)

print(randseed1.randn(5))
print(randseed1.randn(5))
print(randseed2.randn(5))
print(randseed1.randn(5))
```

## pytorch

```python
import torch
```

```python
torch.randn(5)
```

```python
torch.manual_seed(10)
print(torch.randn(5))

print(np.random.randn(5))
```
