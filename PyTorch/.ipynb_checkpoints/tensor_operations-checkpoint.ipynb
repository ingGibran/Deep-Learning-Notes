# Tensor Operations

## Matrix Multiplication vs Element-wise

```python
import torch 

A = torch.tensor([
    [1, 2],
    [3, 4]
])

B = torch.tensor([
    [5, 6],
    [7, 8]
])

print(f"Element-Wise: {A * B}")
print(f"Matmul: {A @ B}")
```

## View vs reshape

* .view Returns the same memory
* .reshape Returns the values

```python
x = torch.rand(2, 3)

flat = x.view(6)
print(flat.shape)

new_shape = x.reshape(3, 2)
print(new_shape.shape)
```

## Batch Dimension

```python
img = torch.rand(3, 224, 224)

img_batch = img.unsqueeze(0)
print(img_batch.shape)

img_orig = img_batch.squeeze(0)
print(img_orig.shape)
```
