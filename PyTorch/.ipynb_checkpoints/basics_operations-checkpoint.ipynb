## Basic Operations

```python
import torch 
import numpy as np
```

## Importing and creating

```python
x_data = [1, 2, 3]

ts = torch.tensor(x_data)
print(ts)

shape = (2, 3)

# Random
rand_tensor = torch.rand(shape)
print(rand_tensor)

# Ones
ones_tensor = torch.ones(shape)
print(ones_tensor)

# Zeros
zeros_tensor = torch.zeros(shape)
print(zeros_tensor)
```

## Load data

```python
# numpy -> pytorch
np_array = np.array([10, 20, 30])
tensor = torch.from_numpy(np_array)

# pytorch -> numpy
back_to_numpy = tensor.numpy()

tensor.add_(1)
print(tensor)
print(np_array)
```
